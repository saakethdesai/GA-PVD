```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
from matplotlib import font_manager

font_dirs = ['/Users/saadesa/Downloads/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

# set font
plt.rcParams['font.family'] = 'CMU Sans Serif'
```

```python
timestep_minusone = np.loadtxt("1e-1.txt", skiprows=2)
timestep_minustwo = np.loadtxt("1e-2.txt", skiprows=2)
timestep_minusthree = np.loadtxt("1e-3.txt", skiprows=2)
timestep_minusfour = np.loadtxt("1e-4.txt", skiprows=2)
```

```python
timestep_minustwo[:, -1]
```

```python
%matplotlib notebook

plt.figure(figsize=(10, 10))

plt.plot(timestep_minustwo[:, -1]/1000, lw=4, c="#d95f02")
plt.plot(timestep_minusthree[:, -1]/1000, lw=4, c="#7570b3")
plt.plot(timestep_minusfour[:, -1]/1000, lw=4, c="#1b9e77")

plt.yticks([-8, -6, -4, -2, 0])
plt.xlabel("Time(steps)*timestep", fontsize=24)
plt.ylabel("Free energy (x 10$^{3}$)", fontsize=24)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
#plt.xlim([0, 130])
```

```python
%matplotlib notebook

plt.figure(figsize=(11, 10))

base = timestep_minusthree[:, -1]

plt.plot((timestep_minustwo[:,-1] - base)*100/base, lw=4, c="#d95f02")
plt.plot((timestep_minusthree[:,-1] - base)*100/base, lw=4, c="#7570b3")
plt.plot((timestep_minusfour[:,-1] - base)*100/base, lw=4, c="#1b9e77")

#plt.yticks([-8, -6, -4, -2, 0])
plt.xlabel("Time(steps)*timestep", fontsize=24)
plt.ylabel("Percentage error in free energy", fontsize=24)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
#plt.xlim([0, 130])
```
